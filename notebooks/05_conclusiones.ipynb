## 1. Respuesta a las preguntas de análisis

### Pregunta 1 — ¿El plan de suscripción está asociado a un mayor consumo mensual?

- **Evidencia:** en `03_eda.ipynb`, el consumo promedio mensual crece de forma escalonada según el
  plan: 597 min (Básico) → 872 min (Estándar) → 1.140 min (Premium). En `04_pca.ipynb`, la
  componente PC2 (dominada casi exclusivamente por el consumo) también ordena a los usuarios de la
  misma forma.
- **Interpretación:** dos fuentes de evidencia independientes (EDA y PCA) muestran el mismo patrón,
  lo que da más confianza en que no es casualidad.
- **Conclusión:** sí, el consumo mensual está asociado al plan de suscripción, y es la relación más
  fuerte encontrada en todo el proyecto.

### Pregunta 2 — ¿La cantidad de tickets de soporte varía según el plan contratado?

- **Evidencia:** el promedio de tickets es prácticamente igual en los tres planes (entre 0.79 y
  0.81 tickets).
- **Interpretación:** no hay una diferencia que sugiera que un plan tenga mejor o peor soporte
  percibido, ni que los usuarios de un plan determinado tengan más problemas técnicos.
- **Conclusión:** no encontramos evidencia de que el plan influya en la cantidad de tickets de
  soporte generados.

### Pregunta 3 — ¿Existe relación entre la edad y el tiempo de consumo?

- **Evidencia:** la correlación entre `age` y `monthly_watch_time_mins` es prácticamente nula
  (≈ 0.004), sin ninguna tendencia visible en el gráfico de dispersión.
- **Interpretación:** usuarios de todas las edades consumen, en promedio, cantidades similares de
  contenido.
- **Conclusión:** no encontramos evidencia de que la edad explique el consumo en este dataset.

### Pregunta 4 — ¿La distribución de planes es similar en todos los países?

- **Evidencia:** las proporciones de cada plan son casi idénticas en los 7 países (~45% Básico,
  ~35% Estándar, ~20% Premium en todos los casos).
- **Interpretación:** no se observan preferencias regionales distintas en la elección de plan.
- **Conclusión:** el país no parece ser un factor relevante para explicar el plan elegido, al menos
  con la información disponible.

### Pregunta 5 — De las variables numéricas disponibles, ¿cuál se asocia más al plan?

- **Evidencia:** en el EDA, solo el consumo mensual mostró una relación clara con el plan; en el
  PCA, la varianza quedó repartida casi por igual entre las 3 componentes (ninguna domina), pero la
  única que sí ordena a los planes es justamente la dominada por el consumo (PC2).
- **Interpretación:** el consumo es, por lejos, la variable más informativa del dataset para
  entender el comportamiento del usuario.
- **Conclusión:** el consumo mensual es la variable numérica más asociada al plan de suscripción;
  la edad y los tickets de soporte no aportan una relación equivalente.

## 3. Limitaciones

- El alcance de las conclusiones se encuentra condicionado por la información disponible y por las
  decisiones documentadas durante el proceso (por ejemplo, la resolución de `user_id` duplicados
  con datos conflictivos, donde no había forma de saber cuál registro era el correcto).
- El dataset presenta un pequeño grupo de edades muy bajas (0 y 4 años, 46 registros) que no se
  trató como error por no contar con una regla clara para distinguirlo de un dato válido; esto
  pudo introducir un sesgo menor en los análisis relacionados con la edad.
- Las variables disponibles son limitadas (no hay información de dispositivo, tiempo de sesión,
  motivo de cancelación, etc.), lo que restringe la profundidad de las conclusiones sobre el
  comportamiento del usuario.
- El PCA se calculó imputando nulos puntuales con la mediana solo para ese análisis, lo cual puede
  suavizar levemente las relaciones reales entre variables.
